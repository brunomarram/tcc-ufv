```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
def removeDelimiters(tweets):
    tweets = tweets.replace("]", "")
    tweets = tweets.split("', '")
    tweets[0] = tweets[0][1:]
    tweets[-1] = tweets[-1][:-1]
    tweets = tweets[0].split('","')
    return tweets
```

```python
def splitTweets(tweets):
    return tweets.split("\n");
```

```python
def splitTweet(tweet):
    stringDelimiter = tweet.split("[")
    latitude = stringDelimiter[0].split(",")[0]
    longitude = stringDelimiter[0].split(",")[1]
    arrayOfTweets = removeDelimiters(stringDelimiter[1])
    return latitude, longitude, arrayOfTweets
```

```python
def writeDistribuition(rawFile):
    try:
        f = open("/code/src/distribuicoes/distribuitionRelated.csv", "a")
    except:
        f = open("/code/src/distribuicoes/distribuitionRelated.csv", "w+")
    f.write(rawFile)
    f.close()
```

```python
def processFile(date, rawTweets, dictionary):
    tweets = splitTweets(rawTweets)
    
    try:
        for splitedTweet in tweets:
            latitude, longitude, arrayOfTweets = splitTweet(splitedTweet)
            arrayOfTweets = [i for i in arrayOfTweets if i] #remove tweets em branco

            try:
                dictionary[latitude+','+longitude][date] = len(arrayOfTweets)
            except:
                dictionary[latitude+','+longitude] = { date: len(arrayOfTweets) }

            print(latitude, longitude, arrayOfTweets)
    except:
        print("Erro")
        
    return dictionary
```

```python
def getHeader(files):
    rawFile = "latitude,longitude,"
    
    for date in files:
        if(date) != files[-1]:
            rawFile += date + ','
        else:
            rawFile += date + '\n'
            
    return rawFile
```

```python
def getColumns(dates, order):
    rawLine = ","
    for date in order:
        try:
            rawLine += str(dates[date]) + ','
        except:
            rawLine += '0,'
    return rawLine[:-1]
```

```python
files = ['16-06', '22-06', '29-06', '06-07', '13-07', '20-07', '10-08', '18-08', '24-08', '01-09', '07-09', '15-09', '21-09']

rawFile = getHeader(files)
    
dictionary = {}
    
for date in files:
    tweetsFile = open('/code/src/normalized/relationed/'+date+'.csv', 'r')
    rawTweets = tweetsFile.read()
    dictionary = processFile(date, rawTweets, dictionary)
    
for place in dictionary:
    rawFile += place
    rawFile += getColumns(dictionary[place], files)
    rawFile += "\n"
    
writeDistribuition(rawFile)
```
