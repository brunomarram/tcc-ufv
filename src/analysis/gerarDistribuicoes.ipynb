```python
import json
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
def removeDelimiters(tweets):
    tweets = tweets.replace("]", "")
    tweets = tweets.split("', '")
    tweets[0] = tweets[0][1:]
    tweets[-1] = tweets[-1][:-1]
    tweets = tweets[0].split('","')
    return tweets
```

```python
def parseObject(string):
    string = string.replace("True", "'True'")
    string = string.replace("False", "'False'")
    arrayOfTweets = eval('['+string)
    texts = []
    for tweet in arrayOfTweets:
        tweet['text'] = tweet['text'].replace("\n", " ")
        texts.append(tweet['text'])
    return texts
```

```python
def splitTweets(tweets):
    return tweets.split("\n");
```

```python
def splitTweet(tweet):
    stringDelimiter = tweet.split("[", 1)
    latitude = stringDelimiter[0].split(",")[1]
    longitude = stringDelimiter[0].split(",")[2]
    arrayOfTweets = parseObject(stringDelimiter[1])
    return latitude, longitude, arrayOfTweets
```

```python
def writeDistribuition(rawFile):
    try:
        f = open("data/distribuicao-tweets/distribuition-v2.csv", "a")
    except:
        f = open("data/distribuicao-tweets/distribuition-v2.csv", "w+")
    f.write(rawFile)
    f.close()
```

```python
def processFile(date, rawTweets, dictionary):
    tweets = splitTweets(rawTweets)
    
    for splitedTweet in tweets:
        try:
            latitude, longitude, arrayOfTweets = splitTweet(splitedTweet)
            arrayOfTweets = [i for i in arrayOfTweets if i] #remove tweets em branco

            try:
                dictionary[latitude+','+longitude][date] = len(arrayOfTweets)
            except:
                dictionary[latitude+','+longitude] = { date: len(arrayOfTweets) }

            print(latitude, longitude, len(arrayOfTweets))
        except Exception as e:
            print(f"Erro {e}")
        
    return dictionary
```

```python
def getHeader(files):
    rawFile = "latitude,longitude,"
    
    for date in files:
        if(date) != files[-1]:
            rawFile += date + ','
        else:
            rawFile += date + '\n'
            
    return rawFile
```

```python
def getColumns(dates, order):
    rawLine = ","
    for date in order:
        try:
            rawLine += str(dates[date]) + ','
        except:
            rawLine += '0,'
    return rawLine[:-1]
```

```python
files = ['10-08', '18-08', '24-08', '01-09', '07-09', '15-09', '21-09']

rawFile = getHeader(files)
    
dictionary = {}
    
for date in files:
    tweetsFile = open('../data-manipulation/tweets/v2/hospitals/'+date+'.csv', 'r')
    rawTweets = tweetsFile.read()
    dictionary = processFile(date, rawTweets, dictionary)
    
for place in dictionary:
    rawFile += place
    rawFile += getColumns(dictionary[place], files)
    rawFile += "\n"
    
writeDistribuition(rawFile)
```
