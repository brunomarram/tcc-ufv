```python
import pandas as pd
import numpy as np
from datetime import datetime
import re
from collections import Counter
import json
import ast
```

```python
place = "Hospitais"
stopwords = pd.read_csv('./stopwords.txt')
```

```python
def removeDelimiters(tweets):
    tweets = tweets.replace("]", "")
    tweets = tweets.split("', '")
    tweets[0] = tweets[0][1:]
    tweets[-1] = tweets[-1][:-1]
    return tweets
```

```python
def splitTweets(tweets):
    return tweets.split("\n");
```

```python
def parseObject(string):
    string = string.replace("True", "'True'")
    string = string.replace("False", "'False'")
    arrayOfTweets = eval('['+string)
    texts = []
    for tweet in arrayOfTweets:
        tweet['text'] = tweet['text'].replace("\n", " ")
        texts.append(tweet['text'])
    return texts
```

```python
def splitTweet(tweet):
    stringDelimiter = tweet.split("[", 1)
    latitude = stringDelimiter[0].split(",")[1]
    longitude = stringDelimiter[0].split(",")[2]
    arrayOfTweets = parseObject(stringDelimiter[1])
    return latitude, longitude, arrayOfTweets
```

```python
def removeStopwords(tweet):
    for stopword in stopwords['stopwords']:
        stopword = stopword.replace(" ", "")
        tweet = tweet.replace(" "+stopword+" ", " ")
        tweet = tweet.replace(" "+stopword+"…", "…")
        tweet = tweet.replace(" "+stopword+".", ".")
        tweet = tweet.replace(" "+stopword+",", ",")
    return tweet
```

```python
def removeNumerals(tweet):
    #tweet = re.sub('[^a-zA-Z0-9 \n\.]', ' ', tweet)
    tweet = re.sub('[0-9]', '', tweet)
    return tweet
```

```python
def splitLink(tweet):
    splited = tweet.split("https")
    tweet = splited[0]
    link = None
    if len(splited) > 1:
        link = 'https'+splited[1]
    return tweet, link
```

```python
def normalizeTweet(tweet):
    tweet, link = splitLink(tweet)
    tweet = removeStopwords(tweet)
    tweet = removeNumerals(tweet)
    return tweet, link
```

```python
def isRelevant(tweet):
    relevantWords = [
        " dor ", " dores ", "dolorido", "sentindo mal", "senti mal", "covid", "corona",
        "febre", "tosse", "cansado", "cansaço", "garganta", "diarreia", "conjutivite", "paladar",
        "ouvido", "esfolação", "isolado", "trancado", "fiqueemcasa", "pandemia", "isolamentosocial",
        "saude", "quarentena", "RT-PCR", "cotonete", "teste", "internado", "exame", "imunocromatografia",
        "sorologia", "imune"
    ]
    for word in relevantWords:
        if(word in tweet):
            return True
    return False
```

```python
def writeTweets(filename, line):
    try:
        f = open("./tweets-relationed-v2.1/"+filename+".csv", "a")
    except:
        f = open("./tweets-relationed-v2.1/"+filename+".csv", "w+")
    f.write(line + '\n')
    f.close()
```

```python
def processFile(date, rawTweets):
    tweets = splitTweets(rawTweets)

    count = 0
    
    for splitedTweet in tweets:
        try:
            latitude, longitude, arrayOfTweets = splitTweet(splitedTweet)
            line = latitude+','+longitude+','+'['

            for rawTweet in arrayOfTweets:
                try:
                    tweet, link = normalizeTweet(rawTweet)
                    if isRelevant(tweet) and link is not None:
                        count += 1
                        line += '"'+tweet+'",'
                except:
                    print("não consegui processar o tweet: " + tweet)
            if(line[-1] == ','):
                line = line[:-1] + ']'
            else:
                line += ']'

            writeTweets(date, line)
            print("linha escrita")
        except Exception as e:
            print("Finalizado"+ str(e))
            
    print("Resultado de "+str(count)+" tweets relevantes em "+date)
    return count
```

```python
files = ['24-05', '31-05', '07-06', '14-06', '21-06', '28-06', '05-07']

total = 0

for date in files:
    fileToNormalize = 'v3/hospitals/' + date
    saveAt = "hospitals/" + date

    tweetsFile = open('../tweets/'+fileToNormalize+'.csv', 'r')
    rawTweets = tweetsFile.read()
    total += processFile(date, rawTweets)
    
print("Total de tweets relevantes em "+ place + ": "+str(total))
```
